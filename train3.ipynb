```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = []
with open('wpbc.data', 'r') as f:
    for line in f:
        line = list(map(float, line.replace('R', '1').replace('N', '0').replace('?', '1.54').split(',')))
        line = line[:2] + [x / 100 for x in [line[2], line[3], line[5], line[18], line[20], line[24],
                           line[25], line[27], line[28], line[31]]] 
        data.append(line)           
```

```python
x, y = [], []
t = int(len(data) * 0.8)
for i in data[:t]:
    x.append(i[2:])
    y.append(int(i[1]))
x
```

```python
x = np.matrix(x)
y = np.matrix(y).transpose()
y.shape
```

```python
w1 = np.random.rand(10, 20)
w2 = np.random.rand(20, 1)
```

```python
learning_rate = 0.5
loss = []
n = len(x)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
for _ in range(1000):
    z1 = x.dot(w1)
    l1 = sigmoid(z1)
    z2 = l1.dot(w2)
    y_hat = sigmoid(z2)
    d1 = x.transpose().dot(np.multiply(np.multiply(y_hat - y, 1 - y_hat, y_hat).dot(w2.transpose()), l1, 1 - l1))/n
    d2 = l1.transpose().dot(np.multiply(y_hat, 1 - y_hat, y_hat - y)) / n
    w1 = w1 - learning_rate * d1
    w2 = w2 - learning_rate * d2
    L = np.multiply(y_hat - y, y_hat - y).sum() / (2 * n)
    loss.append(L)
    print(L)
    
    
```

```python

from matplotlib import pyplot as plt
plt.plot(loss)
plt.show()
```
